```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Відобразимо створений xls - документ, що містить дані про марки автомобілів, їхню вартість, споживання палива, максимальну швидкість та вантажопідйомність:

```python
df = pd.read_excel('C:/Users/Vitalii/Downloads/cars_set.xlsx', index_col=0)
df
```

**Візуалізуємо наступні залежності:**

*а) споживання палива від вартості лінійною діаграмою:*

```python
fig, ax = plt.subplots()
s = df['Вартість авто, $']
plt.plot(s.sort_values(ascending=True), df['Споживання палива, л/100км'], color = 'green', linewidth = 3)
ax.set_xlabel('Вартість авто, $')
ax.set_ylabel('Споживання палива, л/100км')
```

*б) споживання палива від швидкості точковим графіком:*

```python
fig, ax = plt.subplots()
plt.scatter(df['Макс. швидкість км/год'], df['Споживання палива, л/100км'], marker = 'D')
ax.set_xlabel('Макс. швидкість км/год')
ax.set_ylabel('Споживання палива, л/100км')
```

*в) вантажопідйомності від швидкості точковим графіком:* 

```python
fig, ax = plt.subplots()
plt.scatter(df['Макс. швидкість км/год'], df['Вантажопідйомність, кг'], marker = '*')
ax.set_xlabel('Макс. швидкість км/год') 
ax.set_ylabel('Вантажопідйомність, кг') 
```

**Побудуємо гістограму для розділення автомобілів на 5 груп за швидкістю:**

```python
fig, ax = plt.subplots()
plt.hist(df['Макс. швидкість км/год'], 5, color = 'green')
ax.set_xlabel('Макс. швидкість км/год')
ax.set_ylabel('Кількість')
```

**Побудуємо коробковий графік для відображення швидкостей різних марок:**

```python
plt.subplots()
sns.boxplot(x = df['Марка авто'], y = df['Макс. швидкість км/год'], data = df)
```

**Побудуємо парний графік для відображення залежностей величин:**

```python
sns.pairplot(df)
```
